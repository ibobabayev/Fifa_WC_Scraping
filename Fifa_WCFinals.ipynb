```python
from bs4 import BeautifulSoup
import requests
import pandas
```

```python
url = "https://en.wikipedia.org/wiki/List_of_FIFA_World_Cup_finals"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html')
```

```python
table = soup.find('table',class_="sortable plainrowheaders wikitable")

headers = table.find_all('th')

headers_name = [data.text.strip() for data in headers][:7]

data_frame = pandas.DataFrame(columns=headers_name)

data_in_rows = table.find_all('tr')
```

```python
for row in data_in_rows[1:-1]:
    th_data = row.find_all('th')[0].find('a').contents[0]
    data = row.find_all('td')[:-1]
    individual_data = [item.text.strip() for item in data]
    individual_data.insert(0,th_data)

    length = len(data_frame)
    data_frame.loc[length] = individual_data
```

```python
data_frame = data_frame.astype({'Year': 'int32'})
data_frame.dtypes
```

```python
data_frame['Attendance']=data_frame['Attendance'].str.replace(',','',regex=True)
data_frame
```

```python
data_frame = data_frame.astype({'Attendance': 'int32'})
data_frame.dtypes
```

```python
#Year of Final with the most Attendance
data_frame.loc[data_frame['Attendance'].idxmax(),'Year']
```

```python
#Finals with the most Attendance`s
data_frame.sort_values(by='Attendance',ascending=False)
```

```python
#Countries with the most World Cup`s
data_frame.groupby('Winners')[['Year']].count().sort_values(by='Year',ascending=False)
```

```python
#Countries with the most losses in World Cup finals
data_frame.groupby('Runners-up')[['Year']].count().sort_values(by='Year',ascending=False)
```

```python
#Finals that have gone to extra time 
data_frame[data_frame['Score'].str.contains('a.e.t')]
```

```python
#Finals that have gone to the penalty shoot-out 
data_frame[data_frame['Score'].str.contains('pen')]
```

```python
data_frame.to_csv('Fifa_WCFinals.csv', sep='\t', encoding='utf-8', index=False, header=True)
```
